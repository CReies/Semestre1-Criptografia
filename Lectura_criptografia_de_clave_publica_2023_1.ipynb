####Problema 2. Implementación del algoritmo de exponenciación modular $\color{green}{\text{(2 puntos)}}$
---

A continuación, se muestra una posible implementación del ALGORITMO 5 en Python. Estudia el código y explica qué hace cada línea. **Ten en cuenta el límite de palabras**.

```python
b = 232; n = 79; m = 3337

a = bin(n)

x = 1
potencia = b % m

for i in range(len(a)-2):
  if int(a[-1-i]) == 1:
    x = (x * potencia) % m
  potencia = (potencia * potencia) % m

print(x)
```
